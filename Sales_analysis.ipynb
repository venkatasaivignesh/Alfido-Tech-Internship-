```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales = pd.read_csv("/content/sample_data/superstore_final_dataset.csv",  encoding='latin-1')
```

```python
sales.head(10)
```

```python
sales.sample(20)
```

# **Scrubing**
1. Removing the first column, there's already row values present.

```python
sales = sales.iloc[:,1:]
sales.head()
```

2. Filling Null Values.

```python
null = sales['Postal_Code'].isna()
sales.loc[null,'Postal_Code'] = 0
sales.Postal_Code.isna().value_counts()
```

```python
sales.isna().value_counts()
```

3. Removing Duplicate Values


```python
sales.duplicated().value_counts()
```

```python
dup = sales.duplicated(keep = "first")
sales = sales.loc[-dup,:]
```

```python
sales.duplicated().value_counts()
```

# **Analyzing**
Now, performing analysis on this sales dataset.

```python
sales.describe()
```

```python
sales["Order_Date"] = pd.to_datetime(sales["Order_Date"],"%d %m %Y")
sales["Year"] = pd.DatetimeIndex(sales["Order_Date"]).year
sales.head()
```

```python
sales_year = sales.groupby('Year')['Sales'].sum()
sales_year
```

```python
cx = sales_year.plot(kind = "line", color = "violet")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.title("Sales w.r.t Year")

for x in cx.containers:
  cx.bar_label(x)
```

Sales increased from year 2016, to sky high.

```python
region = sales.groupby('Region')['Sales'].sum()
region
```

```python
bx = region.plot(kind = "bar",color = 'brown')
plt.xlabel("Region")
plt.ylabel("Sales")
plt.title("Sales w.r.t Region")

for x in bx.containers:
  bx.bar_label(x)
```

From the West Region, the Sales were higher, compared to other regions.

```python
df= sales['Category'].value_counts()
df
```

```python
ax = df.plot(kind='bar')
plt.xlabel("Category")
plt.ylabel("Count")
plt.title("No. of Sales w.r.t Category")

for x in ax.containers:
    ax.bar_label(x)
plt.show()
```

```python
df2 = sales.groupby('Category')['Sales'].sum()
df2
```

```python
ax1 = df2.plot(kind='bar', color = 'red')
plt.xlabel("Category")
plt.ylabel("Sales")
plt.title("Sales w.r.t Category")

for x in ax1.containers:
    ax1.bar_label(x)
plt.show()
```

Here, Technology Category has most sales compared to other.

```python
df3 = sales.groupby('Sub_Category')['Sales'].sum()
df3 = df3.sort_values(ascending=False)
df3
```

```python
category = sales.groupby('Category')['Sub_Category'].value_counts()
category
```

```python
ax = df3.plot(kind='bar', color = 'green')
plt.xlabel("Sub_Category")
plt.ylabel("Sales")
plt.title("Sales w.r.t Sub_Category")

for x in ax.containers:
    ax.bar_label(x)
```

In Technology Category, customers are mostly spend in phones.

```python
df4 = sales.groupby('Sub_Category')['Sales'].mean()
df4 = df4.sort_values(ascending=False)
df4
```

```python
ax2 = df4.plot(kind='bar', color = 'orange')
plt.xlabel("Sub_Category")
plt.ylabel("Average Sales")
plt.title("Average Sales w.r.t Sub_Category")
for x in ax2.containers:
  ax2.bar_label(x)
```

When it comes to sales, Copiers from Technology Category, has the highest average sale rate.

## Convert 'order date' to datetime

### Subtask:
Convert the 'Order_Date' column to datetime objects to enable easy extraction of the year.


**Reasoning**:
Convert the 'Order_Date' column to datetime objects using the specified format.



```python
sales['Order_Date'] = pd.to_datetime(sales['Order_Date'], format='%d/%m/%Y')
sales.info()
```

## Extract year from 'order date'

### Subtask:
Create a new column 'Order_Year' containing the year from the 'Order_Date' column.


**Reasoning**:
Create a new column 'Order_Year' containing the year from the 'Order_Date' column by accessing the year attribute of the datetime objects.



```python
sales['Order_Year'] = sales['Order_Date'].dt.year
sales.head()
```
